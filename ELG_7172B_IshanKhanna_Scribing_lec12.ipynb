## Aleatory and Epistemic Uncertainties

There are two classes of uncertainities:

Aleatory Uncertainity : is uncertainity that comes from random process of which we are observing such as:
 
a) *Flipping coin and predicting heads and tails*

b) *Drawing card from the deck*

c) *parameter that is being measured such as RPM of motor , pressure in turbulent boundary layer.* 

Epistematic uncertainities : Uncertainities that arise due to lack of knowledge of paramaters of the computational model and model errors such as 

a) mechanical properties of material result because of change of scale or

b) geometrical tolerances induced by the manufacturing process 

that are not releated to the parameters of model.

# Major Challenges for the Computational Models

## 1) Which robustness must be looked for computational model

In most of the cases, it is not adequate to have detreministic computational models of complex real systems. This can be analyzed early by comparing the predictions with experimental data. The robustness can be improved by considering uncertainities in computational model for which it is always possible to make stochastic model with the help of mathematical properties of uncertain paramaters, variabilities of real system and experimental errors due to noise in measurement or lack of knowledge of measured configuration.

## 2) Why the probability theory and the mathematical statistics are efficient

Probability theory is used to build and determine the prior and the posterior stochastic models of uncertainties in finite or infinite dimension, for examining propagation of uncertainities with help of applied mathemathics, for solving statistical inverse problems and make it possible for model to learn from data.

## 3) What Types of Stochastic Analyses Can Be Done

To perform in uncertain computational model, The analyses like Robust predictions, design optimization and performance optimization can be done with respect to uncertainities, variablities and experimental errors even if the data is not available. If data is given then apart from above mentioned analysis, robust updating can also be performed.


## 4) Uncertainty quantification and model validation must be carried out


Stochastic computational model should be build through developing stochastic model of uncertainities for given computational model and then use it to do propagation of uncertainities. Uncertainities must quantified with the help of data (if given), stochastic computational model and tools that solve statistical inverse problems.


## What are the major challenges

1) Development of stochastic model of uncertainities that contain all the information to make model better, even when data is not given. For random vectors and tensor-valued random fields in high stochastic dimension and random matrices in any stochastic dimension, we must develop alzebric representations, random generators and probability distribution which is very difficult to construct.

2) Another challenging problem is to do stochastic modeling of model uncertainities when data is not given.

3) One of the problem is to update and identification of stochastic model of uncertainities by solving statistical inverse problem if limited data is available and it will be more challenging if model is in high stochastic dimension.

4) The last challenge is algorithm strategy for robust optimization, robust design and robust updating. It requires clever interpolation methods and knowledge of reduced order models.

## Inverse Problem

Inverse problem : It is basically to indetify some parameters of computational model by using observed data of quantities of interest that can be calculated with the computational model. It is called inverse problem because it identifies cause from the result. 


Statistical inverse problems : Inverse problems that can be solved in probabilistic framework by mathematical statistical tools.
These Inverse problem can be solved if data is given. The data can be either obtained from numerical simulations with computational models or experimental observations. 

ILL-posed problem : If data is not available for statistical inverse problem then problem is called ill-posed problem that can not be solved.

Well-posed problem : If data is available for statistical inverse problem then problem is called well-posed problem and solution quality depends on the quality of the convergence of the statistical estimates with respect to the size of data.

There are two methods to solve well-posed inverse problem. one is least square method and second method is bayesian model.

##  Least Square Method [2]

In this example, We will solve inverse problem $G=md$ (where G is data kernel, m is model parameter and d is our data) by using least square method and will measure the length of the estimated model parameter $m^{est}$ and of predicted data $d^{pre}=Gm^{est}$. It uses measures of length as the best practice in solving an inverse problem by fitting straight line to data.

In least square method, we always try to choose model parameters in such a way that our predicted data is always close to the observed data and computes prediction error for each observation. Lets consider the model $Gm=d$ depends on the model paramater $m^{est}$ and predicted data $d^{pre}=Gm^{est}$. The overall error $E$ is 

$$\begin{equation*}
E = \sum_{i=1}^N e_i^2 = \mathbf{e}^{\text{T}} \mathbf{e}
\end{equation*}$$

where e is $\mathbf{e} = \mathbf{d}^{\text{obs}}-\mathbf{d}^{\text{pre}}$

$$\begin{equation*}
E = \mathbf{e}^T \mathbf{e} = (\mathbf{d} - \mathbf{Gm})^T(\mathbf{d}-\mathbf{Gm}) = \sum_{i=1}^N \left[ d_i - \sum_{j=1}^M G_{ij}m_j \right] \left[ d_i - \sum_{k=1}^M G_{ik}m_k \right]
\end{equation*}$$

$$\begin{equation*}
E = \sum_{j=1}^M \sum_{k=1}^M m_j m_k \sum_{i=1}^N G_{ij} G_{ik} - 2 \sum_{j=1}^M m_j \sum_{i=1}^N G_{ij}d_i + \sum_{i=1}^N d_i d_i
\end{equation*}$$

After computing the derivative of E with one of the parameter $m_q$,
$$\partial E / \partial m_q = 2 \sum_{k=1}^M m_k \sum_{i=1}^N G_{iq} G_{ik} -2 \sum_{i=1}^N G_{iq} d_i + 0 $$

By setting the result equal to 0, we get

$$0 = 2 \sum_{k=1}^M m_k \sum_{i=1}^N G_{iq} G_{ik} - 2 \sum_{i=1}^N G_{iq} d_i$$

Writing this equation in matrix notation yields

$$\mathbf{G}^T \mathbf{Gm} - \mathbf{G}^T \mathbf{d} = 0 $$
$$\mathbf{m}^{\text{est}} = [\mathbf{G}^T \mathbf{G}]^{-1} \mathbf{G}^T \mathbf{d}$$

which is the least squares solution to the inverse problem Gm=d.







```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
# STRAIGHT LINE PROBLEM: Least squares fit to synthetic data

# x-axis synthetic data
z = np.sort(np.random.uniform(0, 10, 30))

# observed synthetic data (d = a + bz + noise)
dobs = 2.0 + 1.0 * z + np.random.normal(0, 1.0, 30)

# data kernel
M = 2
G = np.array([np.ones(30), z])

# standard matrix solution
mest1 = inv(G.dot(np.transpose(G))).dot(G.dot(dobs)).reshape(2,1)
dpre1 = np.transpose(G).dot(mest1)

plt.subplots(1,1, figsize=(7,5))

# PLOT 1: Simulated data and least squares solution
plt.subplot(111)

# simulated data
plt.scatter(z, dobs, color="black")

# least squares solution
plt.plot(z, dpre1, color="blue")

# graphing params
plt.xlim(0,10)
plt.ylim(0,15)
plt.title('Least Squares Fitting')
plt.xlabel('z')
plt.ylabel('d')
plt.show()
```

We have our fitted model parameters but we need to check the quality of that fit. 

## Bayesian Method [3]

In  Bayesian method, the solution to an inverse problem is the probability distribution of the quantity of interest, i.e. the unknown parameters.  

lets perform above experiment (G=md) N times and every time $d_i$ is recorded and data is all noisy measurements of the same model parameter $m_1$.As we know, All N realizations will have the same probability density function. If these random variables are Gaussian, their joint probability density function will be

$$\begin{equation*}
p(\mathbf{d}) = \sigma^{-N} (2 \pi)^{-N/2} \: \text{exp} \left[ - \frac{1}{2} \sigma^{-2} \sum_{i=1}^N [d_i - m_1]^2 \right]\\
\end{equation*}$$
The data $\mathbf{d}^{\text{obs}}$ and probability density function can be represented graphically as a point in the N-dimensional space.Our expected result is as all the d's have the same mean and variance, probability density function will be centered around line $d_1 = d_2 = \cdots = d_N$.

```python
import math

import numpy as np
from numpy.linalg import inv
from numpy.linalg import det

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from skimage import measure

plt.style.use('ggplot')
%matplotlib inline

# point inside a 3d box

x = np.arange(31) * 5 / 30
y = np.arange(31) * 5 / 30
z = np.arange(31) * 5 / 30

# randomly choose a point in 3-space from a normal distribution (mean=2.5, sd=0.5)
rbar1 = np.random.normal(2.5, 0.5, 3)

# build 3d grid of probability density function
rbar2 = np.array([(2.5, 2.5, 2.5)])
sd = 0.5
C = sd**2 * np.eye(3)
CI = inv(C)
DC = det(C)
norm = (2 * math.pi)**(3/2) * DC**0.5

# calculate PDF probabilities
PP = np.zeros((31,31,31))
for i in range(0,31):
    for j in range(0,31):
        for k in range(0,31):
            r = np.array([(i * 5/30, j * 5/30, k * 5/30)])
            PP[i,j,k] = math.exp((-0.5 * (r-rbar2))[0].dot(CI.dot((r-rbar2).reshape(3,1))))/norm

# threshold PDF probability (PP) volume as a triangulated isosurface
verts, faces,a,b = measure.marching_cubes(PP, np.max(PP)/10, spacing=(5/30, 5/30, 5/30))
            
# build figure
plt.subplots(1, 2, figsize=(15, 7))

# randomly chosen point graph
ax = plt.subplot(121, projection='3d')
ax.scatter(rbar1[0], rbar1[1], rbar1[2])
plt.plot(x,y,z, color='black')
ax.set_xlim3d(0, 5)
ax.set_ylim3d(0, 5)
ax.set_zlim3d(0, 5)
ax.view_init(10, 225)
plt.title('Realization of a random variable')

# PDF as a spherical cloud graph
ax = plt.subplot(122, projection='3d')
plt.plot(x,y,z, color='black')
ax.plot_trisurf(verts[:, 0], verts[:,1], faces, verts[:, 2], cmap='Spectral', lw=0)
ax.set_xlim3d(0, 5)
ax.set_ylim3d(0, 5)
ax.set_zlim3d(0, 5)
ax.view_init(10, 225)
plt.title('Probability density function as spherical cloud')
plt.show()
```

##  Bayes’ Method for Estimating the Posterior Probability from a Prior Probability Model


Bayesian methods allow us to estimate posterior pdf $p_x^{Post}(x)$ of random vaiable X for which a prior pdf $p^{prior}_X(x)$ and $ν$ experimental data of number of realizations should be available and the random vector $U^{out}$ is introduced such that
$U^{out} = g(X,B)$ to compare observation with measured quantity. If there is additive noise B then
$$U^{out}=h(X)+B$$

According to bayes law,
$$p^{post}_X(x) = c_n L(x) p^{prior}_X (x)$$ 

where $L(x)$ is likelihood function. We are interested in computing $E\{q(U)\}$ which is  
$$E\{q(U)\}=E\{q(h(X))\} =\int_R q(h(x)) p^{post}_X(x)dx$$

MCMC methods can used to compute the integeral of high dimensions.






In the below code, we choose our model as $y=\alpha + \beta X + \epsilon $ where $\epsilon$ is additive noise. Here we are trying to estimate $ \alpha$ , $\beta$ and $\epsilon$ by MAP. First, we have defined the prior distribution for our estimate paramaters. <br>

'Y_obs = pm.Normal('Y_obs', mu=mu, sd=elipsion, observed=Y)'- This represents the likelihood of the data under model. Again we define the variable name and set parameter values with mu and sd. The parameter mu and elipsion is assigned to these random variables, indicating that we are trying to model that variables. Lastly, we provide observed instances of the variable (i.e. our data) with the observed keyword. Because we have said this variable is observed, the model will not try to change its values.

Third, we generate samples of estimate paramaters from the posterior using the sampling algorithm and starting values. Here we have used No-U-Turn sampling but one can choose other algorithms such as Metropolis, Hastings, Gibbs, and Slice sampling.

```python
import pymc3 as pm
import numpy as np

alpha, epsilon = 1, 1
beta = 1

# Size of dataset
size = 100

# Predictor variable
X1 = np.random.randn(size)

# Simulate outcome variable
Y = alpha + beta*X1  + np.random.randn(size)*epsilon
        
with pm.Model() as model:

# Set prior 
    alpha = pm.Normal('alpha', mu=0, sd=10)
    beta = pm.Normal('beta', mu=0, sd=10)
    epsilon = pm.HalfNormal('epsilon', sd=1)

    # Expected value of outcome
    mu = alpha + beta*X1 

    # Likelihood (sampling distribution) of observations
    Y_obs = pm.Normal('Y_obs', mu=mu, sd=epsilon, observed=Y)
    start = pm.find_MAP()     #Max Aposterior values, or values that are most likely

    step = pm.NUTS()     #Choose a particular MCMC algorithm     #we'll choose NUTS, the No U-Turn Sampler (Hamiltonian)

    trace = pm.sample(5000, step, start=start, progressbar=True)
```

```python
print(pm.summary(trace))
pm.traceplot(trace)
```

Results of above program will provide full probability distribution of various values of our estimated parameters, we can take the mean of the distribution as our most plausible value for our parameters, which is 1.01 for $\alpha$, 1.18 for $\beta$ and 1.04 for $\epsilon$.
